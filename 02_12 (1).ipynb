```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

first = pd.read_csv('products.csv')
second = pd.read_csv('order_details.csv')

#a = first['UnitPrice']
#b = second['Quantity']

first.head()
```

```
happy = second.groupby('ProductID').sum() #it was grouped acording to ProductID
sad = pd.merge(happy, first, on = 'ProductID')  #product id of first and second csv's are merged
sad.head()
#matplotlib
```

```
fig = plt.figure() #to create a figure 
axes = fig.add_axes([0.1,0.1,0.9,0.9]) #changes the size of the figure (left, bottom, width, height)
axes.scatter(sad['UnitPrice_x'], sad['Quantity']) #creates a scatter plot figure where x=unitprice x and y= quantity
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pk_df = pd.read_csv('Pokemon.csv')

glob = pk_df[((pk_df['Type 1'] == 'Fighting') | (pk_df['Type 2'] == 'Fighting')) | (pk_df['Type 1'] == 'Normal') | (pk_df['Type 2'] == 'Normal')]
#30 pokemons (y) with 60 defense (x)
fig2 = plt.figure()
axes = fig2.add_axes([0.1,0.1,0.9,0.9])
axes.hist(glob['Defense'], bins = 30, alpha = 0.9, rwidth = 0.9)
```
